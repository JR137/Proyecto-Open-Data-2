<p style="color: darkblue">Primero instalamos(pip install tika-app) e importamos la libreria que vamos a utilizar </p>

```python
from tikapp import TikaApp
```

<p style="color: darkblue">La clase TikaReader encapsula la funcionalidad para interactuar con la biblioteca Tika en Python. Su constructor inicializa una instancia de TikaApp con la ruta proporcionada al archivo JAR de Tika. Los métodos de la clase permiten detectar el tipo de contenido y el idioma de un archivo, así como extraer todo su contenido o solo el texto. Además, la función load_tika_jar proporciona una forma conveniente de cargar el archivo JAR de Tika. En conjunto, estas herramientas facilitan el procesamiento de documentos y la extracción de información utilizando Tika en entornos Python.</p>

```python
class TikaReader:
    def __init__(self, jar_path):
        self.tika_app = TikaApp(jar_path)
    def detect_content_type(self, file_path):
        return self.tika_app.detect_content_type(file_path)
    def detect_language(self, file_path):
        return self.tika_app.detect_language(file_path)
    def extract_all_content(self, file_path, convert_to_obj=False):
        return self.tika_app.extract_all_content(file_path, convert_to_obj=convert_to_obj)
    def extract_only_content(self, file_path):
        return self.tika_app.extract_only_content(file_path)
def load_tika_jar(jar_path):
    return TikaApp(jar_path)
```

<p style="color: darkblue">La función process_document toma la ruta de un archivo como entrada y utiliza la clase TikaReader para realizar varias operaciones en el archivo utilizando la biblioteca Tika. En primer lugar, crea una instancia de TikaReader con la ruta al archivo JAR de Tika. Luego, utiliza los métodos de esta instancia para detectar el tipo de contenido y el idioma del archivo, así como para extraer todo el contenido y solo el texto del archivo. Finalmente, devuelve una tupla que contiene el tipo de contenido, el idioma, todo el contenido y el contenido solo de texto del archivo. Esta función proporciona una forma conveniente de procesar documentos y extraer información utilizando Tika en Python.</p>

```python
def process_document(file_path):
    tika_reader = TikaReader('tika-app-1.20.jar')
    content_type = tika_reader.detect_content_type(file_path)
    language = tika_reader.detect_language(file_path)
    all_content = tika_reader.extract_all_content(file_path)
    only_content = tika_reader.extract_only_content(file_path)
    return content_type, language, all_content, only_content
```

<p style="color: darkblue">Este fragmento de código utiliza la función process_document para procesar un archivo PDF específico llamado 'EjerciciosTPA_JhosuaCallejas.pdf'. Luego imprime los resultados obtenidos, incluyendo el tipo de contenido, el idioma, todo el contenido extraído y el contenido de texto solamente.</p>

```python
file_path = 'EjerciciosTPA_JhosuaCallejas.pdf'
content_type, language, all_content, only_content = process_document(file_path)

print("Content Type:", content_type)
print("Language:", language)
print("All Content:", all_content)
print("Only Content:", only_content)
```

<p style="color: darkblue">La clase ProcessJSONTika proporciona funcionalidades para procesar el JSON generado por Tika al analizar un archivo. En su constructor, recibe una instancia de TikaReader, permitiendo así la comunicación con la biblioteca Tika para extraer información del archivo.</p>

<p style="color: darkblue">El método process_json toma la ruta de un archivo como entrada y utiliza el TikaReader para extraer todo el contenido del archivo en formato JSON. Luego, comprueba si se obtuvo algún dato del archivo. Si es así, accede al primer elemento del JSON (asumiendo que la lista devuelta contiene un solo elemento) y extrae metadatos como el autor, la fecha de creación y la fecha de modificación. También imprime el contenido del archivo si está disponible. Si no se pudo extraer ningún dato del archivo, se imprime un mensaje indicando este hecho.</p>

<p style="color: darkblue">Este enfoque proporciona una forma de procesar archivos utilizando Tika y obtener información útil, como metadatos y contenido, en formato JSON.</p>

```python
class ProcessJSONTika:
    def __init__(self, tika_reader):
        self.tika_reader = tika_reader

    def process_json(self, file_path):
        json_data = self.tika_reader.extract_all_content(file_path, convert_to_obj=True)
        if json_data:
            json_data = json_data[0]
            print("Metadatos:")
            print("Autor:", json_data.get('Author', 'No disponible'))
            print("Fecha de Creación:", json_data.get('Creation-Date', 'No disponible'))
            print("Fecha de Modificación:", json_data.get('Last-Modified', 'No disponible'))
            print()
            content = json_data.get('X-TIKA:content', 'No disponible')
            print("Contenido del archivo:")
            print(content)
        else:
            print("No se pudo extraer ningún dato del archivo.")
```

<p style="color: darkblue">La función main sirve como punto de entrada principal para ejecutar el procesamiento de un archivo PDF utilizando las clases TikaReader y ProcessJSONTika. En primer lugar, se crea una instancia de TikaReader inicializada con la ruta al archivo JAR de Tika ('tika-app-1.20.jar'). Luego, se crea una instancia de ProcessJSONTika, pasando la instancia de TikaReader como argumento. A continuación, se define la ruta del archivo PDF a procesar ('EjerciciosTPA_JhosuaCallejas.pdf'). Finalmente, se llama al método process_json de ProcessJSONTika con la ruta del archivo PDF como argumento, lo que desencadena el procesamiento del archivo y la impresión de metadatos y contenido.</p>

<p style="color: darkblue">El bloque if __name__ == "__main__": asegura que la función main se ejecute solo si este script se ejecuta directamente y no se importa como un módulo en otro script. Esto es una buena práctica en Python para modularizar el código y mantener la flexibilidad de reutilización.</p>

```python
def main():
    tika_reader = TikaReader('tika-app-1.20.jar')
    json_processor = ProcessJSONTika(tika_reader)
    file_path = 'EjerciciosTPA_JhosuaCallejas.pdf'
    json_processor.process_json(file_path)
if __name__ == "__main__":
    main()
```
