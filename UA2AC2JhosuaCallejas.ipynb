```python
from tikapp import TikaApp
```

<p style="color: darkblue">La clase TikaReader encapsula la funcionalidad para interactuar con la biblioteca Tika en Python. Su constructor inicializa una instancia de TikaApp con la ruta proporcionada al archivo JAR de Tika. Los mÃ©todos de la clase permiten detectar el tipo de contenido y el idioma de un archivo, asÃ­ como extraer todo su contenido o solo el texto. AdemÃ¡s, la funciÃ³n load_tika_jar proporciona una forma conveniente de cargar el archivo JAR de Tika. En conjunto, estas herramientas facilitan el procesamiento de documentos y la extracciÃ³n de informaciÃ³n utilizando Tika en entornos Python.</p>

```python
class TikaReader:
    def __init__(self, jar_path):
        self.tika_app = TikaApp(jar_path)
    def detect_content_type(self, file_path):
        return self.tika_app.detect_content_type(file_path)
    def detect_language(self, file_path):
        return self.tika_app.detect_language(file_path)
    def extract_all_content(self, file_path, convert_to_obj=False):
        return self.tika_app.extract_all_content(file_path, convert_to_obj=convert_to_obj)
    def extract_only_content(self, file_path):
        return self.tika_app.extract_only_content(file_path)
def load_tika_jar(jar_path):
    return TikaApp(jar_path)
```

<p style="color: darkblue">La funciÃ³n process_document toma la ruta de un archivo como entrada y utiliza la clase TikaReader para realizar varias operaciones en el archivo utilizando la biblioteca Tika. En primer lugar, crea una instancia de TikaReader con la ruta al archivo JAR de Tika. Luego, utiliza los mÃ©todos de esta instancia para detectar el tipo de contenido y el idioma del archivo, asÃ­ como para extraer todo el contenido y solo el texto del archivo. Finalmente, devuelve una tupla que contiene el tipo de contenido, el idioma, todo el contenido y el contenido solo de texto del archivo. Esta funciÃ³n proporciona una forma conveniente de procesar documentos y extraer informaciÃ³n utilizando Tika en Python.</p>

```python
def process_document(file_path):
    tika_reader = TikaReader('tika-app-1.20.jar')
    content_type = tika_reader.detect_content_type(file_path)
    language = tika_reader.detect_language(file_path)
    all_content = tika_reader.extract_all_content(file_path)
    only_content = tika_reader.extract_only_content(file_path)
    return content_type, language, all_content, only_content
```

<p style="color: darkblue">Este fragmento de cÃ³digo utiliza la funciÃ³n process_document para procesar un archivo PDF especÃ­fico llamado 'EjerciciosTPA_JhosuaCallejas.pdf'. Luego imprime los resultados obtenidos, incluyendo el tipo de contenido, el idioma, todo el contenido extraÃ­do y el contenido de texto solamente.</p>

```python
file_path = 'EjerciciosTPA_JhosuaCallejas.pdf'
content_type, language, all_content, only_content = process_document(file_path)

print("Content Type:", content_type)
print("Language:", language)
print("All Content:", all_content)
print("Only Content:", only_content)
```

<p style="color: darkblue">La clase ProcessJSONTika proporciona funcionalidades para procesar el JSON generado por Tika al analizar un archivo. En su constructor, recibe una instancia de TikaReader, permitiendo asÃ­ la comunicaciÃ³n con la biblioteca Tika para extraer informaciÃ³n del archivo.</p>

<p style="color: darkblue">El mÃ©todo process_json toma la ruta de un archivo como entrada y utiliza el TikaReader para extraer todo el contenido del archivo en formato JSON. Luego, comprueba si se obtuvo algÃºn dato del archivo. Si es asÃ­, accede al primer elemento del JSON (asumiendo que la lista devuelta contiene un solo elemento) y extrae metadatos como el autor, la fecha de creaciÃ³n y la fecha de modificaciÃ³n. TambiÃ©n imprime el contenido del archivo si estÃ¡ disponible. Si no se pudo extraer ningÃºn dato del archivo, se imprime un mensaje indicando este hecho.</p>

<p style="color: darkblue">Este enfoque proporciona una forma de procesar archivos utilizando Tika y obtener informaciÃ³n Ãºtil, como metadatos y contenido, en formato JSON.</p>

```python
class ProcessJSONTika:
    def __init__(self, tika_reader):
        self.tika_reader = tika_reader

    def process_json(self, file_path):
        json_data = self.tika_reader.extract_all_content(file_path, convert_to_obj=True)
        if json_data:
            json_data = json_data[0]
            print("Metadatos:")
            print("Autor:", json_data.get('Author', 'No disponible'))
            print("Fecha de CreaciÃ³n:", json_data.get('Creation-Date', 'No disponible'))
            print("Fecha de ModificaciÃ³n:", json_data.get('Last-Modified', 'No disponible'))
            print()
            content = json_data.get('X-TIKA:content', 'No disponible')
            print("Contenido del archivo:")
            print(content)
        else:
            print("No se pudo extraer ningÃºn dato del archivo.")
```

<p style="color: darkblue">La funciÃ³n main sirve como punto de entrada principal para ejecutar el procesamiento de un archivo PDF utilizando las clases TikaReader y ProcessJSONTika. En primer lugar, se crea una instancia de TikaReader inicializada con la ruta al archivo JAR de Tika ('tika-app-1.20.jar'). Luego, se crea una instancia de ProcessJSONTika, pasando la instancia de TikaReader como argumento. A continuaciÃ³n, se define la ruta del archivo PDF a procesar ('EjerciciosTPA_JhosuaCallejas.pdf'). Finalmente, se llama al mÃ©todo process_json de ProcessJSONTika con la ruta del archivo PDF como argumento, lo que desencadena el procesamiento del archivo y la impresiÃ³n de metadatos y contenido.</p>

<p style="color: darkblue">El bloque if __name__ == "__main__": asegura que la funciÃ³n main se ejecute solo si este script se ejecuta directamente y no se importa como un mÃ³dulo en otro script. Esto es una buena prÃ¡ctica en Python para modularizar el cÃ³digo y mantener la flexibilidad de reutilizaciÃ³n.</p>

```python
def main():
    tika_reader = TikaReader('tika-app-1.20.jar')
    json_processor = ProcessJSONTika(tika_reader)
    file_path = 'EjerciciosTPA_JhosuaCallejas.pdf'
    json_processor.process_json(file_path)
if __name__ == "__main__":
    main()
```
